### Queries using PySpark DSL

DSL = Domain Specific Language - API similar to natural or other language, implemented as library in another language.

List all the countries with the population > 38 million

```python
countries.filter((col("Population") > 38000000)).orderBy("Population").toPandas()
```

Select all the countries from Europe

```python
countries.select("Country", "Population").where(col("Region").like("%EUROPE%")).show()
```

Conditions in `where` clause can contain logical expressions.

```python
countries.select("Country", "Population")\
.where((col("Region").like("%EUROPE%")) & (col("Population")> 10000000)).show()
```

## User defined functions for data manipulation
Our `countries` DataFrame has some problems:
 * missing values
 * some numbers use comma instead of point as floating point separator (e.g. Literacy  = 99,4)

We can clean the data using User Defined Functions (UDF)

```python
from pyspark.sql.types import FloatType
from pyspark.sql.functions import udf
```

Define a Python function which coverts numbers with commas to `float`

```python
def to_float (s) :
   return float(s.replace(',','.'))
```

Test that it works:

```python
to_float('0,99')
```

Now define a Spark UDF:

```python
float_udf = udf(to_float , FloatType())
```

Test it on a Data Frame

```python
countries.withColumn("Literacy", float_udf("Literacy (%)"))
```

OK, we can see that the `Literacy` is now `float`

```python
countries.show(50)
```

```python
countries.where((col("Literacy (%)") < 50) & (col("GDP ($ per capita)") > 700)).show()
```

Oops, what does it mean???

- some rows have empty values!

Before we can use the table, we need to remove empty rows. Otherwise our UDF will fail.

```python
full_countries = countries.select('Country', 'Population', 'Literacy (%)', 'GDP ($ per capita)').na.drop()
```

We can now apply the new UDF to the Data Frame:

```python
full_countries = full_countries.withColumn("Literacy", float_udf("Literacy (%)"))
```

```python
full_countries.show(50)
```

```python
full_countries.where((col("Literacy") < 50) & (col("GDP ($ per capita)") > 700)).show()
```

```python
full_countries.toPandas().plot(x="Literacy",y="GDP ($ per capita)",kind="scatter",figsize=(10, 6))
```

# Narysuj wykres zależności kolumn Birthrate i Deathrate w zależności od GDP per capita dla 30 najbogatszych i 30 najbiedniejszych krajów.

```python
from pyspark.sql.functions import lit
full_countries = countries.select('Country', 'Birthrate', 'Deathrate', 'Literacy (%)', 'GDP ($ per capita)').na.drop()

# Filter countries with 30 highest and 30 lowest GDP
top_countries = full_countries.orderBy(col("GDP ($ per capita)").desc()).limit(30)
bottom_countries = full_countries.orderBy(col("GDP ($ per capita)").asc()).limit(30)


```

```python
top_df = top_countries.toPandas()
bottom_df = bottom_countries.toPandas()

# Create scatter plot for Birthrate vs GDP
plt.figure(figsize=(10, 6))

# Plot top countries in red
plt.scatter(top_df["GDP ($ per capita)"], top_df["Birthrate"], color='red', label='Top 30 GDP Countries')

# Plot bottom countries in blue
plt.scatter(bottom_df["GDP ($ per capita)"], bottom_df["Birthrate"], color='blue', label='Bottom 30 GDP Countries')

# Adding titles and labels
plt.title('Birthrate vs GDP ($ per capita)')
plt.xlabel('GDP ($ per capita)')
plt.ylabel('Birthrate')

# Adding legend to differentiate between top and bottom countries
plt.legend()

# Show the plot
plt.show()
```

```python
# Create scatter plot for Birthrate vs GDP
plt.figure(figsize=(10, 6))

# Plot top countries in red
plt.scatter(top_df["GDP ($ per capita)"], top_df["Deathrate"], color='red', label='Top 30 GDP Countries')

# Plot bottom countries in blue
plt.scatter(bottom_df["GDP ($ per capita)"], bottom_df["Deathrate"], color='blue', label='Bottom 30 GDP Countries')

# Adding titles and labels
plt.title('Deathrate vs GDP ($ per capita)')
plt.xlabel('GDP ($ per capita)')
plt.ylabel('Deathrate')

# Adding legend to differentiate between top and bottom countries
plt.legend()

# Show the plot
plt.show()
```

# Znajdź 10 krajów w których najniżej położone lotnisko ma najwyższą wysokość, podając wyniki w m. n. p. m.

Czyli 10 krajów które mają tylko 1 lotnisko??

```python
from pyspark.sql.functions import col, concat, lit

#  countries with exactly 1 airport
countries_with_one_airport = airports.groupBy("country") \
    .count() \
    .filter(col("count") == 1) \
    .select("country")

one_airport_countries = airports.join(countries_with_one_airport, "country", "inner")
one_airport_countries= one_airport_countries.select("country", "name", col("altitude").cast("int").alias("altitude"))

#  10 airports with the highest altitude
top_10_altitude = one_airport_countries.orderBy((col("altitude").cast("int")).desc()).limit(10)


# Display the result
top_10_altitude.select("country", "name", "altitude").orderBy(col("altitude").desc()).show()
```

# Na wykresie scatterplot narysuj zależność liczby lotnisk od powierzchni kraju.
Dane potrzebne do wykresu są w dwóch ramkach (Countries i Airports), konieczne więc będzie ich złączenie (join).
Nie zawsze nazwy tych samych krajów będą identyczne w obu ramkach, co może stanowić problem (złączenie nie powstanie). Zastanów się jak przy pomocy mechanizmu złączeń wykryć, które nazwy się różnią. Stwórz odpowiednią ramkę, która zawiera takie nazwy.


```python
from pyspark.sql.functions import count, col, trim
import seaborn as sns
import matplotlib.pyplot as plt


airports_count = airports.groupBy("country").agg(count("*").alias("airport_count"))
countries = countries.withColumn("Country", trim(countries["Country"]))
joined_df = airports_count.join(countries, airports_count["country"] == countries["Country"])

data_for_plot_pd = joined_df.toPandas()

data_for_plot_pd.head()
```

```python
# Rysowanie wykresu scatterplot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data_for_plot_pd, x="Area (sq. mi.)", y="airport_count")
plt.title("")
plt.xlabel("Area")
```

# Oblicz, ile lotnisk jest na poszczególnych kontynentach. To zadanie również wymaga złączenia ramek "Countries" i "Airports". Wynik przedstaw na wykresie kołowym.

```python
airports_per_continent = joined_df.groupBy("Region").agg(count("airport_count").alias("total_airports"))

airports_per_continent_pd = airports_per_continent.toPandas()


plt.figure(figsize=(8, 8))
plt.pie(airports_per_continent_pd["total_airports"], labels=airports_per_continent_pd["Region"], autopct='%1.1f%%', startangle=140)

plt.show()
```
